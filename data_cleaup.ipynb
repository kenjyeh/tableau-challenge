```python
df = pd.read_csv("Resources/2021_citibike.csv")
df.head()
```

```python
#keep records with longer than 90 seconds and if ending station isn't the same as the start station
df = df.loc[((df["tripduration"]>90) & (df["start station id"] != df["end station id"])),:]
df.head()
```

```python
# convert trip duration to minutes
df["tripduration"] = [trip/60 for trip in df["tripduration"]]
df.rename(columns={"tripduration":"tripduration (min)"})
df.head()
```

```python
# change gender column to strings
gender = {0:"Unknown", 1:"Male", 2:"Female"}
df.gender = [gender[id] for id in df.gender]
df.head()
```

```python
# Calculate age of riders
ages = []

for i in df["birth year"]:
    age = 2021 - i
    ages.append(age)
    

# Append new Age column
df["Age"] = ages
df
```

```python
#set age bins
bins = [0, 17, 25, 35, 50, 65, 90]
labels = ["0-17", "18-25", "26-35", "36-50", "51-65", "65-90"]

df["Age Group"] = pd.cut(df["Age"], bins, labels=labels)

del df["birth year"]
```
